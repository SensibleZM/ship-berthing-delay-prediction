```python
# ===================== 船舶延误数据集自动生成代码=====================
import pandas as pd
import numpy as np

# 1. 设定随机种子，保证结果可复现
np.random.seed(42)

# 2. 定义数据集核心字段
# 训练集8000条样本，测试集2000条样本，符合真实比赛/项目的数据比例
train_samples = 8000
test_samples = 2000

# 类别特征枚举值
ship_types = ['集装箱船', '散货船', '油轮', '杂货船', '滚装船', '邮轮']
port_of_origins = ['上海港', '宁波舟山港', '深圳港', '青岛港', '天津港', '新加坡港', '釜山港', '鹿特丹港']
arrival_terms = ['高峰', '平峰']
service_years = ['<1年', '1-3年', '3-5年', '5-10年', '10-15年', '15年以上']

# 3. 生成训练集
train_data = {
    # 船舶属性特征
    'Ship_Type': np.random.choice(ship_types, size=train_samples, p=[0.35, 0.25, 0.15, 0.1, 0.1, 0.05]),
    'Gross_Tonnage': np.random.randint(1000, 250000, size=train_samples),
    'Years_In_Service': np.random.choice(service_years, size=train_samples, p=[0.05, 0.15, 0.2, 0.3, 0.2, 0.1]),
    # 航行特征
    'Voyage_Distance': np.random.randint(100, 12000, size=train_samples),
    'Port_Of_Origin': np.random.choice(port_of_origins, size=train_samples),
    'Arrival_Term': np.random.choice(arrival_terms, size=train_samples, p=[0.4, 0.6]),
    # 环境特征
    'Wind_Speed': np.round(np.random.uniform(0, 28, size=train_samples), 1),
    'Visibility': np.round(np.random.uniform(0.1, 10, size=train_samples), 1),
}

# 生成标签Delay_Risk（0=准点，1=延误），符合真实业务逻辑：风速越高、吨位越大、高峰时段，延误概率越高
train_df = pd.DataFrame(train_data)
delay_prob = (
    0.15 + 
    (train_df['Wind_Speed'] / 30) * 0.4 + 
    (train_df['Gross_Tonnage'] / 250000) * 0.2 + 
    (train_df['Arrival_Term'] == '高峰') * 0.15 +
    (train_df['Visibility'] < 1) * 0.2
)
train_df['Delay_Risk'] = np.random.binomial(1, delay_prob.clip(0, 0.95), size=train_samples)

# 4. 生成测试集
test_data = {
    'Ship_Type': np.random.choice(ship_types, size=test_samples, p=[0.35, 0.25, 0.15, 0.1, 0.1, 0.05]),
    'Gross_Tonnage': np.random.randint(1000, 250000, size=test_samples),
    'Years_In_Service': np.random.choice(service_years, size=test_samples, p=[0.05, 0.15, 0.2, 0.3, 0.2, 0.1]),
    'Voyage_Distance': np.random.randint(100, 12000, size=test_samples),
    'Port_Of_Origin': np.random.choice(port_of_origins, size=test_samples),
    'Arrival_Term': np.random.choice(arrival_terms, size=test_samples, p=[0.4, 0.6]),
    'Wind_Speed': np.round(np.random.uniform(0, 28, size=test_samples), 1),
    'Visibility': np.round(np.random.uniform(0.1, 10, size=test_samples), 1),
}
test_df = pd.DataFrame(test_data)

# 5. 生成提交模板sample_submission
submission_df = pd.DataFrame({
    'id': range(test_samples),
    'Delay_Risk': 0
})

# 6. 保存为csv文件，和项目代码路径完全匹配
train_df.to_csv('train.csv', index=False, encoding='utf-8-sig')
test_df.to_csv('test.csv', index=False, encoding='utf-8-sig')
submission_df.to_csv('sample_submission.csv', index=False, encoding='utf-8-sig')

# 打印验证信息，确认生成成功
print("✅ 数据集生成完成！")
print(f"训练集形状：{train_df.shape}")
print(f"测试集形状：{test_df.shape}")
print(f"提交模板形状：{submission_df.shape}")
print("===== 训练集标签分布 =====")
print(train_df['Delay_Risk'].value_counts())
print("文件已保存到当前文件夹，可直接运行项目代码！")
```
